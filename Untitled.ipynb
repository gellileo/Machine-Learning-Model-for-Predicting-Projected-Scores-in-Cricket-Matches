```python
df=pd.get_dummies(data=df,columns=["batting_team","bowling_team"])
```

```python
df.columns
```

```python
x=df[['batting_team_Australia','batting_team_Bangladesh', 'batting_team_England', 'batting_team_India', 'batting_team_New Zealand', 'batting_team_Pakistan','batting_team_South Africa', 'batting_team_Sri Lanka',
       'batting_team_West Indies', 'bowling_team_Australia',
       'bowling_team_Bangladesh', 'bowling_team_England', 'bowling_team_India',
       'bowling_team_New Zealand', 'bowling_team_Pakistan',
       'bowling_team_South Africa', 'bowling_team_Sri Lanka','bowling_team_West Indies', 'over_number','ball_number',
     'total_score', 'total_wickets']]
y=df["total"].values
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.20)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
#from sklearn.ensemble import RandomForestClassifier 
from sklearn import linear_model
RFR = linear_model.Ridge(alpha=50, max_iter=100, tol=0.1)


```

```python
#RFR=RandomForestClassifier(n_estimators=100)
```

```python

RFR.fit(x_train,y_train)
```

```python
y_pred=RFR.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
err=mean_absolute_error (y_test, y_pred)
err
```

```python
#function to predict scores
def predict_score(bat_team,bowl_team,over_number,ball_number,total_score,total_wickets):
    temp=list()
    if bat_team=="Australia":
        temp=temp+[1,0,0,0,0,0,0,0,0]
    elif bat_team=="Bangladesh":
        temp=temp+[0,1,0,0,0,0,0,0,0]
    elif bat_team=="England":
        temp=temp+[0,0,1,0,0,0,0,0,0]
    elif bat_team=="India":
        temp=temp+[0,0,0,1,0,0,0,0,0]
    elif bat_team=="New Zealand":
        temp=temp+[0,0,0,0,1,0,0,0,0]
    elif bat_team=="Pakistan":
        temp=temp+[0,0,0,0,0,1,0,0,0]
    elif bat_team=="South Africa":
        temp=temp+[0,0,0,0,0,0,1,0,0]
    elif bat_team=="Sri Lanka":
        temp=temp+[0,0,0,0,0,0,0,1,0]
    else: 
        temp=temp+[0,0,0,0,0,0,0,0,1]
        
    if bowl_team=="Australia":
        temp=temp+[1,0,0,0,0,0,0,0,0]
    elif bowl_team=="Bangladesh":
        temp=temp+[0,1,0,0,0,0,0,0,0]
    elif bowl_team=="England":
        temp=temp+[0,0,1,0,0,0,0,0,0]
    elif bowl_team=="India":
        temp=temp+[0,0,0,1,0,0,0,0,0]
    elif bowl_team=="New Zealand":
        temp=temp+[0,0,0,0,1,0,0,0,0]
    elif bowl_team=="Pakistan":
        temp=temp+[0,0,0,0,0,1,0,0,0]
    elif bowl_team=="South Africa":
        temp=temp+[0,0,0,0,0,0,1,0,0]
    elif bowl_team=="Sri Lanka":
        temp=temp+[0,0,0,0,0,0,0,1,0]
    else: 
        temp=temp+[0,0,0,0,0,0,0,0,1]
    temp=temp+[over_number,ball_number,total_score,total_wickets]
    data=np.array([temp])
    prediction=int(RFR.predict(data))
    print("predicted score : " ,prediction)
    print("predicted score range : " ,prediction-19 ,"to",prediction+19)
        
        
        
        
        
        
bat_team="Australia"
bowl_team="India"
over_number=10
ball_number=3
total_score=80


total_wickets=3

predict_score(bat_team,bowl_team,over_number,ball_number,total_score,total_wickets)
```
